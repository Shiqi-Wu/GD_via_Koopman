Here in this notebook you will go beyond the MNIST dataset, albeit only a little bit. We will use the [Fashion MNIST](https://github.com/zalandoresearch/fashion-mnist) dataset, which is designed to replace MNIST as the new minimal dataset to test vision tasks. 

It is a dataset of Zalando's article imagesâ€”consisting of a training set of 60,000 examples and a test set of 10,000 examples. # Importing the Dataset

Just like MNIST, this dataset is found in the `tf.keras.datasets` module.

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

The following is a dictionary of labels that will map the y_train/y_test labels into words

```python
label_dict = {
    0: 'T-shirt/top',
    1: 'Trouser',
    2: 'Pullover',
    3: 'Dress',
    4: 'Coat',
    5: 'Sandal',
    6: 'Shirt',
    7: 'Sneaker',
    8: 'Bag',
    9: 'Ankle boot',
}
labels_train = [label_dict[u] for u in y_train]
labels_test = [label_dict[u] for u in y_test]
```

```python
y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)
```

```python
ax = sns.heatmap(x_train[0], cmap='Greys')
ax.set_title(f'Label = {labels_train[0]}')
```

```python
print(x_train.shape)
# print(labels_train.shape)
print(x_test.shape)
# print(labels_test.shape)
```

```python
x_train = x_train.reshape(-1, 784) / 255.0
x_test = x_test.reshape(-1, 784) / 255.0
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
print(y_train.shape)
print(y_train[0])
```

# Task 1: Build a Shallow NN for Classification

Build and train a shallow (one-hidden-layer) neural network with 128 hidden nodes to classify these images. Remember to perform appropriate preprocessing steps, such as normalization and one-hot representations.

Evaluate carefully the performance of your training and the final model's performance. Discuss any interesting phenomena you discover.

## Build and train the shallow NN
We constructed a neural network with a single hidden layer containing 128 nodes, utilizing the 'relu' activation function. The cross-entropy loss function was selected and the model was trained for 500 iterations. Initially, the loss function decreased quickly, but it became challenging to further improve accuracy after 200 training epochs. The loss function fluctuated around 0.4.

image.png
image.png

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=10, activation='softmax'))
```

```python
from tensorflow.keras.optimizers import SGD
```

```python
model.compile(loss='categorical_crossentropy', optimizer=SGD(learning_rate=0.25), metrics=['accuracy'])
```

```python
history = model.fit(x=x_train,y=y_train,epochs=500,batch_size=x_train.shape[0],validation_data=(x_test, y_test))
```

```python
import pickle
with open('/content/drive/MyDrive/Colab Notebooks/Course/DSA5204/homework1/history/history_benchmark.txt', 'wb') as f:
  pickle.dump(history.history, f)
```

```python
import pandas as pd
history = pd.DataFrame(history.history)
history.plot(y=['loss', 'val_loss'], title='Loss')
history.plot(y=['accuracy', 'val_accuracy'], title='Accuracy')
```

## Performance depending on number of hidden nodes
To evaluate the influence of hidden nodes in neural networks, we constructed two-layer neural networks with 8, 32, 128, 256 hidden nodes utilizing the 'relu' activation function. The cross-entropy loss function was selected and the model was trained for 500 iterations.

A small number of hidden nodes in a neural network can result in poor classification approximations. For instance, when the number of hidden nodes is 8, the loss function value does not decrease as effectively as it does in a neural network with 128 hidden nodes. In fact, a two-layer neural network with only 32 hidden nodes performed similarly to a network with 128 hidden nodes on this dataset. Increasing the number of hidden nodes further, such as to 256, did not result in a noticeable improvement in performance in practice, despite the theoretical potential for improvement.

image.png

## Performance depending on activation function

Furthermore, the choice of activation function affects the training performance of our models. To assess the impact of activation functions, we constructed neural networks with a single hidden layer containing 128 nodes, and tested the relu, sigmoid, tanh, and leaky relu activation functions. The cross-entropy loss function was selected and the model was trained for 500 iterations.

The results show that the performance of the relu, relu6, and tanh activation functions was similar in this dataset. However, the loss and accuracy of the sigmoid model showed monotonic properties and the training speed was slower in the sigmoid model compared to the others.

*The activation functions:

relu: $\mathrm{relu}(x)=\max (x,0)$

leaky relu: $\mathrm{relu6}(x)=x$ for $x>0$; $\mathrm{relu6}(x)=\delta x$ for $x\leq0$, $Î´=0.3$

sigmoid: $\mathrm{sig}(x)=\frac{1}{1+e^x}$

tanh: $\tanh(x)=\frac{e^x-e^{-x}}{e^x+e^{-x}}$

image.png
image.png

### activation function: tanh

```python
model_tanh = Sequential()
model_tanh.add(Dense(units=128, activation='tanh'))
model_tanh.add(Dense(units=10, activation='softmax'))
model_tanh.compile(loss='categorical_crossentropy', optimizer=SGD(learning_rate=0.25), metrics=['accuracy'])
history_tanh = model_tanh.fit(x=x_train,y=y_train,epochs=500,batch_size=x_train.shape[0],validation_data=(x_test, y_test))
```

## Performance depending on loss functions
Finally, we analyzed the effect of the loss function. The zero-one loss function is challenging to implement in neural networks because it causes the backpropagation algorithm to break due to its discontinuity. In comparison, the cross-entropy (CE) loss function demonstrated superior accuracy performance compared to the mean squared error (MSE) loss function.
